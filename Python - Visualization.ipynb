## Matplotlib

### Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Create a line plot

```python
# define dataset
x = np.arange(0,10,0.1)
y = 2*x + 5

# plot data points
plt.plot(x,y)
plt.show()
```

### Customizing the line plot

```python
# define dataset
x = np.arange(0,10,0.1)
y = 2*x + 5

# cahnge the figure size
fig = plt.figure(figsize = (20,5))

plt.plot(x,y, linewidth = 2.0, linestyle = ":", color = 'r', alpha = 0.5, marker = "o")
plt.title("Line Plot customization")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(["line1"])
plt.grid(True)
plt.show()
```

### Subplots

```python
x = np.arange(0,10,0.1)
y1 = 2*x + 5
y2 = 3*x + 10
y3 = 4*x + 15
y4 = 5*x + 20

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title('Graph1')

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('Graph2')

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('Graph3')

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('Graph4')

plt.show()
```

```python
x = np.arange(0,10,0.1)
y1 = 2*x + 5
y2 = 3*x + 10
y3 = 4*x + 15
y4 = 5*x + 20


plt.subplot(2,4,1)         # 2 rows ,  4 columns , 1 position/index
plt.plot(x,y1)
plt.title('Graph1')

plt.subplot(2,4,2)
plt.plot(x,y2)
plt.title('Graph2')

plt.subplot(2,4,3)
plt.plot(x,y3)
plt.title('Graph3')

plt.subplot(2,4,4)
plt.plot(x,y4)
plt.title('Graph4')

plt.show()
```

### Bar Plot

```python
data = {"apples" : 20 , "mangoes" : 15 , "Lemon" : 30 , "Oranges" : 10}

names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10,5))

plt.bar(names, values)
plt.show()
```

```python
data = {"apples" : 20 , "mangoes" : 15 , "Lemon" : 30 , "Oranges" : 10}

names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10,5))

plt.bar(names, values , color = ['red', 'yellow' , 'yellow' , 'orange'])
plt.show()
```

```python
data = {"apples" : 20 , "mangoes" : 15 , "Lemon" : 30 , "Oranges" : 10}

names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10,5))

plt.barh(names, values , color = ['orange'])
plt.show()
```

### Scatter Plot

```python
a = [10,20,30,40,50,60,70,80]
b = [5,3,2,5,6,1,4,2]
x = [1,2,3,4,3,2,1,2]

fig = plt.figure(figsize = (10,5))

plt.scatter(a,b, s = 200, c = 'g' , marker = 'o' , alpha = 1)
plt.scatter(a,x, s = 300, c = 'r' , marker = '4' , alpha = 1)
plt.savefig("Scatter_plot_Demo.png")

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

plt.scatter(x, y)

# Adjust the scale of the x and y axes
plt.xlim(0, 6)  # Set the x-axis limits from 0 to 6
plt.ylim(0, 25) # Set the y-axis limits from 0 to 25

plt.title('Scatter Plot with Adjusted Axis Scale')

plt.show()
```

```python
import matplotlib.ticker as ticker

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

plt.scatter(x, y)

# Custom tick locations and labels for the x-axis
x_tick_locations = [1, 2, 3, 4, 5]
x_tick_labels = ['A', 'B', 'C', 'D', 'E']

# Set the tick locations and labels for the x-axis
plt.gca().xaxis.set_major_locator(ticker.FixedLocator(x_tick_locations))
plt.gca().xaxis.set_major_formatter(ticker.FixedFormatter(x_tick_labels))

plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot with Customized Axis Scale')

plt.show()
```

### Histograms

```python
numbers = [10,12,16,19,11,20,26,28,30,35,37,45,50,61,64,67,78,79,85,94,95]

plt.hist(numbers , bins = [0,20,40,60,80,100])
plt.show()
```

### Box Plots

```python
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,20,10,5,10,20,30,40,20,45]
data = list([total , order, discount])

plt.boxplot(data)
plt.show()
```

### pie chart

```python
label = ['Dog', 'Cat', 'Wolf', 'Lion']
sizes = [50,45,60,80]

plt.pie(sizes , labels = label , autopct = "%1.1f%%" , shadow = True, explode = (0,0.2,0.1,0))
plt.show()
```

## Seaborn

### Import Library

```python
import seaborn as sns
```

### Import dataset from seaborn

```python
flights = sns.load_dataset("flights")
```

```python
type(flights)
```

```python
flights.head()
```

```python
flights.shape
```

### Create Lineplot

```python
sns.lineplot(data = flights , x = "year" , y = "passengers" , color = "r")
plt.title("MNumber of passengers by year")
plt.show()
```

```python
flights.dtypes
```

```python
flights['month'].unique()
```

```python
flights['year'].unique()
```

```python
flights.shape
```

```python
flights['year'] = flights['year'].astype('category')
```

```python
flights_grouped = flights.groupby("year").sum(numeric_only = True)
```

```python
flights_grouped = flights[['year','passengers']].groupby("year").sum()
```

```python
flights_grouped.head()
```

```python
flights_grouped = flights_grouped.reset_index()
flights_grouped.head()
```

```python
flights_grouped.dtypes
```

```python
sns.barplot(data = flights_grouped , x = 'year' , y = 'passengers')
plt.show()
```

```python
flights_pivot = flights.pivot(index="year" , columns="month", values="passengers")
flights_pivot.head()
```

```python
sns.lineplot(data = flights_pivot ["Jun"])
plt.title("passengers in june by year")
plt.show()
```

```python
plt.figure(figsize = (20,8))
sns.lineplot(data = flights , x = "year" , y= "passengers" , hue = "month")
plt.title("Number og passengers by year on monthly basis")
plt.show()
```
